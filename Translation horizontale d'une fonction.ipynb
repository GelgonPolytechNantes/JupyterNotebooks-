$$f(x)=x^2$$
La fonction $f$ atteint son minimum pour $x=0$

```python
f = x*x
init_printing(use_latex='mathjax')
display(f)
plot(f,(x, -5, 5, 20))
```

$$f(x+1)=(x+1)^2$$
Comme précédemment, la fonction $f$ atteint son minimum quand on argument vaut 0, c.a.d. ici quand $x+1=0$, c.a.d. quand $x=-1$

```python
g = (x+1)*(x+1)
init_printing(use_latex='mathjax')
display(g)
plot(g,(x, -5, 5, 20),ylabel='f(x+1)')
```
