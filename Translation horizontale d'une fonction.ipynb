Translation de l'argument d'une fonction
========================================

Point préliminaire à l'exercice 18 (Transformée de Fourier)
--------------------------------------------------------

```python
from IPython.display import display
from sympy import *
from sympy import fourier_transform, Piecewise
from sympy.plotting import plot
from sympy.abc import x,y

#import sys
%config InlineBackend.figure_formats = {'png', 'retina'} # makes nicer graphics
%matplotlib inline
```


Ce petit notebook illustre la translation de l'argument d'une fonction, qu'on utilise par exemple quand on caractériser l'effet sur la transformée de Fourier (domaine des fréquences) d'une translation dans le domaine direct (c.a.d. domaine du temps, par exemple). On cherche juste à se convaincre par un exemple simple que la représentation graphique de $f(x+a)$ est la représentation graphique de $f$ translatée de $-a$ (et non de $a$).


Prenons une fonction très simple, par exemple 
$$f(x)=x^2$$
La fonction $f$ atteint son minimum pour $x=0$

```python
f = x*x
init_printing(use_latex='mathjax')
display(f)
plot(f,(x, -5, 5, 20))
```

Considérons maintenant un cas particulier simple de translation de l'argument de la fonction :
$$f(x+1)=(x+1)^2$$
Comme précédemment, la fonction $f$ atteint son minimum quand son argument vaut 0. Cela correspond, dans ce second cas, à la situation où $x+1=0$, c.a.d. quand $x=-1$. C'est ce que confirme le graphique ci-dessous.

```python
f = (x+1)*(x+1)
init_printing(use_latex='mathjax')
display(f)
plot(f,(x, -5, 5, 20),ylabel='f(x+1)')
```

En fait, le +1 est le sens de la translation de la __seconde__ fonction vers la __première__.
